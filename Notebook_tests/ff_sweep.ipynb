```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import os
import sys
import time

#allow pickle
import pickle
import pickletools
from pickle import TRUE
from pickletools import uint8
ffdata=np.load("ff_17-51.npy",allow_pickle=True)

plt.figure(figsize=(12,8))
i=0
#plot ff data on a graph
for ff_output in ffdata:
    FF_in=ff_output[0]
    FF_out_PC=ff_output[1]
    mean_PC=np.mean(FF_out_PC,axis=0)
    std_PC=np.std(FF_out_PC,axis=0)
    plt.plot(FF_in,mean_PC,label=i)
    plt.fill_between(FF_in,mean_PC-std_PC,mean_PC+std_PC,alpha=0.2)
    i+=1
plt.legend()
plt.xlabel("Input Frequency (Hz)")
plt.ylabel("Firing Frequency (Hz)")
plt.title("FF PC")
plt.savefig("ff_pc.png")
plt.show()

plt.figure(figsize=(12,8))
i=0
#plot ff data on a graph
for ff_output in ffdata:
    FF_in=ff_output[0]
    FF_out_PV=ff_output[2]
    mean_PV=np.mean(FF_out_PV,axis=0)
    std_PV=np.std(FF_out_PV,axis=0)
    plt.plot(FF_in,mean_PV,label=i)
    plt.fill_between(FF_in,mean_PV-std_PV,mean_PV+std_PV,alpha=0.2)
    i+=1
plt.legend()
plt.xlabel("Input Frequency (Hz)")
plt.ylabel("Firing Frequency (Hz)")
plt.title("FF PV")
plt.savefig("ff_pv.png")
plt.show()


 

```
