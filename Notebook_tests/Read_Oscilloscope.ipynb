```python
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from cProfile import label

Data=np.genfromtxt('Results_Osc/pc_pv_sst_10_21_11_46.csv', delimiter=',')
Data_time=Data[:,0]
Data_PC=Data[:,1]
Data_PV=Data[:,2]
Data_SST=Data[:,3]

plt.plot(Data_time,Data_PC,c='cadetblue',label='PC')
plt.plot(Data_time,Data_PV,c='lightcoral',label='PV')
plt.plot(Data_time,Data_SST,c='sandybrown',label='SST')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Voltage traces')
plt.grid()
plt.legend(loc="upper right", numpoints=1, fontsize=8,markerscale=3)
```
